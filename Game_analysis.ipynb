<p>
In this study, we analyzed a dataset with historical data on game sales, user and expert ratings, genres and platforms. The main task is to find the patterns of the success of the game, which will allow you to bet on a potentially popular product and plan advertising campaigns for the future.

### Conclusion <p>
The dataset is data on game sales for the period from 1980 to 2016. <p> The purpose of the study of this dataset is to obtain an objective picture for the formation of a forecast regarding the future of 2017.
     As a result of the study, it turned out that the typical lifetime of a gaming platform is 9 years. The period from 2015 to 2016 was chosen as the actual period for the study of game sales. By the current (2016) point in time, some platforms showed a near-zero result in sales - such platforms as DS, PSP, Wii were subsequently excluded from the study for the above reason. A group of platforms such as PS3, PSV, Wii are at the final stage of their existence and it is no longer possible to count on high sales, however, a complete absence of games in 2017 for these platforms is hardly expected, which means that sales for these platforms, although they will be reduced, but will not disappear completely.<p>
        
The most promising platforms for the forecast for 2017. viewed:<p>
   PS4 - sales for 2015 - 120 million copies of games, for 2016 - 70 million. <p>
   XOne - 2015 - 60 million copies, 2016 - 25 million.<p>
   3DS - 2015 - 27 million copies, 2016 - 15 million. <p>
   PC - 2015 - 8 million copies, 2016 - 5 million.<p>
   WiiU - 2015 - 8 million copies, 2016 - 5 million.<p>
   PSV - 2015 - 6 million copies, 2016 - 4 million.<p>
   PS3 - 2015 - 16 million copies, 2016 - 3 million.<p>
The PS4 platform is characterized by a larger number of so-called. outliers, values outside the 1.5-fold interquartile range, which tells us about a large number of so-called. "bestsellers", games with big sales. <p>
             The most popular games during the study period from 2015 to 2016 are games in the genres: Action (425 released games for all platforms - more than 100 million copies sold), Shooter (more than 80 games released - about 100 million sold), Role-Playing (about 130 games released - over 50 million sold), Sports (~100 games released with over 60 million sales).<p>
        Top 5 genres to expect high sales from: "Shooter" - median value close to 400k copies of games sold, also for this genre in 75% of cases games were sold with a circulation of up to 1.75 million copies, the "Sports" genre in second place - the median value lies approximately at around 0.15 million copies of games sold, 75% - 0.5 million, "RPG" - 0.15 million copies of games sold by median value, 75% - 0.38 million, "Fighting" - median value - 150 thousand. copies, 75% - 350 thousand copies, "Simulation" - 0.15 million median, 75% - 200 thousand copies.
The most popular gaming platform for the North American and European markets - PS4 - 42% and 57%, respectively, for the same markets, the console that ranks 2nd in terms of sales - XOne - 33% and 19%, respectively. For the Japanese market, the picture is different - the dominant position with 44% is occupied by the 3DS platform. 2nd place in terms of volume is occupied by the PS4 platform. In general, the Japanese market is characterized by a feature of portable "gaming", portable platforms for the period from 2015 to 2016 among the top 5 platforms occupy 60% of the market.<p>
The genre preferences of the SA and European markets are similar in many respects, there are slight differences only in the proportions and desire of a typical European buyer to drive cars, and an American one - to fight virtually. The Japanese market differs from the previous two - the most pronounced tendency to buy games in the genre of Action - 43%, RPG - 25%.<p>
Game sales by age rating:<p>
According to this indicator, the overall distribution pattern for the American and European markets is very similar, with large circulation games sold for an adult audience of more than 40 million copies, more than 35 million copies of sales for the unrated category. The Japanese market has a different distribution pattern - more than 30 million copies of games rated "unknown", for all teenagers - more than 7 million, and the remaining categories up to 5 million each.

#### Historically most games have been released for DS2 and PS2 platforms, but need to check if they are alive (actually I know they are not)

```python
platform_grouped = (df.groupby('platform', as_index=False)
 .agg({'total_sales':['count', sum],'year_of_release':[min,max,]})
 .pipe(lambda x: x.set_axis(x.columns.map('_'.join), axis=1)))
```

```python
platform_grouped['delta_life_platform'] = platform_grouped['year_of_release_max'] - platform_grouped['year_of_release_min']
```

```python
platform_grouped = (platform_grouped.rename
                    (columns={'year_of_release_min': 'first_year_reference', 'year_of_release_max': 'last_year_reference'}))
```

```python
platform_grouped
```

```python
top_ten_platforms = (platform_grouped[(platform_grouped['last_year_reference'] < 2016)
     & (platform_grouped['platform_'] != 'DS')
     & (platform_grouped['platform_'] != 'PC')].sort_values(by='total_sales_sum',
     ascending=False).reset_index(drop=True).loc[:9])
```
